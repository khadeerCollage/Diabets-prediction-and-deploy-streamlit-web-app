```python
#splitting the data
x = diabets_data.drop('Outcome',axis=1)
y = diabets_data['Outcome']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

# svm classifier now play that
# Training the model

```python
classifier = SVC(kernel='linear')
# 
```

```python
classifier.fit(x_train,y_train)
```

# model Evaluation 

```python
classifier_pred_train = classifier.predict(x_train) 
classifier_score = accuracy_score(classifier_pred_train,y_train)
print("the classifier predict of training data :- ",classifier_score)
```

```python
## for testing data
classifier_pred_test = classifier.predict(x_test) 
classifier_score = accuracy_score(classifier_pred_test,y_test)
print("the classifier predict of testing data :- ",classifier_score)
```
